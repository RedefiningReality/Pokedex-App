```python
from IPython.display import Image
import numpy
import torch

model = torch.hub.load("yolov5", 'custom', path="best.pt", source='local',force_reload=True)
model.conf = 0.05

def test_on(file):
    result = model(file)
    df = result.pandas().xyxy[0]
    s = df.get("name")
    if len(s) > 0:
        return s[0]
```
